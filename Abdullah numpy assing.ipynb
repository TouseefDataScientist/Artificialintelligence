Q5: Linear Algebra

Q5.1: Perform matrix multiplication.

```python
A = np.array([[1, 2, 3],
              [4, 5, 6]])
B = np.array([[7, 8],
              [9, 10],
              [11, 12]])
result = A @ B


print(result)
```

Q5.2: Compute the determinant, inverse, and eigenvalues of a matrix.

```python
A = np.array([[1, 6],
              [9, 4]])
determinant = np.linalg.det(A)
inverse = np.linalg.inv(A)
eigenvalues = np.linalg.eigvals(A)
print("Determinant:", determinant)
print("Inverse:")
print(inverse)
print("Eigenvalues:", eigenvalues)
```

Q5.3: Solve a system of linear equations.

```python
A = np.array([[1, 3], [7, 4]])
b = np.array([5, 6])
x = np.linalg.solve(A, b)

print("Solution for x and y:", x)
```

Q6: Advanced Indexing and Slicing

Q6.1: Use Boolean indexing to filter an array.

```python
array = np.array([50, 20, 30, 40, 50, 60])
condition = array > 30
filtered_array = array[condition]

print("Original array:", array)
print("Condition:", condition)
print("Filtered array:", filtered_array)
```

Q6.2: Use fancy indexing to select specific elements from an array.

```python
array = np.array([10, 20, 30, 40, 50, 60])
indices = [0, 2, 4]
selected_elements = array[indices]

print("Selected elements:", selected_elements)
```
