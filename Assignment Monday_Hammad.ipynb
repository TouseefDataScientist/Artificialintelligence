#### Problem 02: “Student Performance Analysis”

```python
# Dataset for student Performance
np.random.seed(42)
num_students = 85

performance_analysis = ({
'Student_ID' : range(1, num_students + 1),
'Physics_Grade' : np.random.randint(50, 101, size= num_students),  
'Computer_Grade' : np.random.randint(50, 101, size= num_students),  
'Attendance': np.random.uniform(60, 100,size= num_students),
'Extracurricular_Activity': np.random.choice(['Sports', 'Music', 'Art', 'None'], num_students)})

students_df = pd.DataFrame(performance_analysis)
students_df['Average_Grade'] = students_df[['Physics_Grade', 'Computer_Grade']].mean(axis=1)
students_df
```

```python
filtered_students = students_df[(students_df['Average_Grade'] > 85) & (students_df['Attendance'] >= 90)]
failing_students = students_df[(students_df[['Physics_Grade', 'Computer_Grade']] < 60).any(axis=1) & (students_df['Extracurricular_Activity'] != 'None')]
print(filtered_students)
print(failing_students) 
```

```python
# Group by extracurricular activity to get average grades
grouped_activity = students_df.groupby('Extracurricular_Activity').agg(Average_Grade=('Average_Grade', 'mean'),
Total_Attendance=('Attendance', 'sum')).reset_index()

# Find the activity with the highest average grade
highest_avg_activity = grouped_activity.loc[grouped_activity['Average_Grade'].idxmax()]
print(highest_avg_activity)
```

```python
# Create a pass/fail column
students_df['Pass_Fail'] = students_df['Average_Grade'].apply(lambda L: 'Pass' if L >= 55 else 'Fail')

# Pass and Fail Counter
pass_fail_counts = students_df['Pass_Fail'].value_counts()
print(pass_fail_counts)
```

```python
# parents income levels counter
income_data = {'Student_ID': range(1, num_students + 1), 
'Parent_Income_Level': np.random.choice(['High', 'Medium', 'Low'], num_students) }
income_df = pd.DataFrame(income_data)
merged_df = pd.merge(students_df, income_df, on='Student_ID')
income_analysis = merged_df.groupby('Parent_Income_Level')['Average_Grade'].mean().reset_index()
print(income_analysis)
```

```python

merged_df ['Grade_Quartile'] = pd.qcut(merged_df['Average_Grade'], 4, labels=['Q1', 'Q2', 'Q3', 'Q4'])

quartile_analysis = merged_df.groupby('Grade_Quartile').agg(Average_Attendance=('Attendance', 'mean'),Average_Grade=('Average_Grade', 'mean')).reset_index() 

for quartile in merged_df['Grade_Quartile'].unique():subset = merged_df[merged_df['Grade_Quartile'] == quartile]
correlation = subset[['Average_Grade', 'Attendance']].corr().iloc[0, 1] 
print("Correlation in {quartile}: {correlation:.2f}")
```
