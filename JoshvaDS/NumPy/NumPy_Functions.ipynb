# Array Creation
## Checking
1. Data type_________________| arr.dtype
2. Shape____________________| arr.shape
3. Size______________________| arr.size
4. Numbers of Dimension_____| arr.ndim


## 1D Array

```python
arr1 = np.array([1,2,3,4])
print("This is our 1st Array:",arr1)
print("1st Array Datatype is:(",arr1.dtype,")")
print("1st Array Shape is:",arr1.shape)
print("Our 1st Array Contains",arr1.size,"numbers of elements.")
print("Our 1st Array is",arr1.ndim,"Dimensional Array.")
```

## 2D Array

```python
arr2 = np.array([[1,2,3,4],[5,6,7,8]])
print("This is our 2nd Array:\n",arr2)
print("2nd Array Datatype is:(",arr2.dtype,")")
print("2nd Array Shape is:",arr2.shape)
print("Our 2nd Array Contains",arr2.size,"numbers of elements.")
print("Our 2nd Array is",arr2.ndim,"Dimensional Array.")
```
