## Q-1 Sales Data Analysis: You have a NumPy array representing the daily sales of a product over a month. Each element in the array represents the sales for a particular day. Compute the total sales, average sales, and the day with the highest sales.


```python
month_sales = (np.random.randint(1,9,30))*1000
print("Pizza Daily Sale list:",month_sales)
```

```python
# Compute the total sales, average sales, and the day with the highest sales.
print("Pizza Total Sales in a Month:",month_sales.sum(),"PKR")
print("Pizza Average Sales per Day in a Month:",month_sales.mean(),"PKR")
max_sales=month_sales.max()
days =(np.where(month_sales == max_sales)[0]+1)
if days.size == 0:
    print("Pizza Sales are same for all days.")
elif days.size > 1:
    print("Pizza Day with Highest Sales are",days.size,".")
    xd=days.size
    while xd>0:
        print("Pizza Day with Highest Sales:",days[xd-1],"Date")
        xd=xd-1
else:
    print("Pizza Day with Highest Sales:",days[0],"Date")
    
```

## Q-7 Traffic Data Analysis: Given a 2D NumPy array where each row represents the number of vehicles passing through a checkpoint in different hours of the day over a week, calculate the total traffic for each day and the busiest hour of the week.



```python
vehicals = np.random.randint(50,100,(24*7))
```

```python
print("Vehicals Passing by Hourly per-day across a week:\n",vehicals.reshape(7,24))
```

### Busiest hour of the day

```python
b = np.max(vehicals.reshape(24,7), axis=0)
ts = b.size
x=1
while ts>0:
    print("Busiest hour of the day",(x),"passing", b[x-1],"vehicals")
    ts=ts-1
    x=x+1
```

### Busiest hour of Week

```python
rdv = vehicals.reshape(1,(24*7))
mvm = np.max(rdv)
w=np.where(vehicals == mvm)[0]
ws = w.size
y=1
while ws>0:
    print("Busiest hour of Week is",w[y-1],"by passing", mvm,"vehicals")
    ws=ws-1
    y=y+1
```
