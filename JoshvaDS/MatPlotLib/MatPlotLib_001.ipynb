```python
x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])
np.random.shuffle(x)
np.random.shuffle(y)
print("Axis-X:\n",x,"\nAxis-Y:\n",y)

title = {'family':'serif','color':'hotpink','size':20}
lable = {'family':'serif','color':'green','size':15}

plt.subplot( 1,2, 1)
plt.plot(x, y)
plt.title("1st Plot Heading", fontdict = lable)
plt.grid(axis="both",color = 'purple', linestyle = '-.', linewidth = 1)

plt.subplot(1,2, 2)
plt.plot(y, x,color = 'b')
plt.title("2nd Plot Heading", fontdict = lable)
plt.grid(axis="both",color = 'green', linestyle = '-.', linewidth = 1)

plt.suptitle("Dashbord Heading", fontdict = title)
plt.show()
```
