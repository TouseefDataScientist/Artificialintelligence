```python
import seaborn as sns
titanic = sns.load_dataset('titanic')
x= titanic["age"]
y= titanic["fare"]
plt.scatter(x, y)
plt.show()
```

```python
x = titanic['age']
y = titanic['fare']

plt.xlabel('Age')
plt.ylabel('Fare')
plt.title('Scatter Plot of Age vs Fare')

category = [plt.Line2D([0], [0], marker='s',
                       c="w",label='200 to Max',
                       markerfacecolor='purple',
                       markersize=10,
                       markeredgecolor='black'),
           plt.Line2D([0], [0], marker='s',
                       c="w",label='70 to 200',
                       markerfacecolor='blue',
                       markersize=10,
                       markeredgecolor='black'),
           plt.Line2D([0], [0], marker='s',
                      c="w",label='0 to 90',
                      markerfacecolor='royalblue',
                      markersize=10,
                      markeredgecolor='black')]

plt.legend(handles=category, title='Fare Categories')
plt.scatter(x, y,c=["purple" if x>200 else "blue" if x<70 else "royalblue" for x in y], edgecolor= "black" )
# plt.colorbar(c=["purple" if x>200 else "blue" if x<70 else "royalblue" for x in y])



plt.show()
```

# Bar Plot

```python
class_counts = titanic['pclass'].value_counts()
class_counts.plot(kind='bar', color=["purple","blue","royalblue"], edgecolor= "black" )
plt.xlabel('Passenger Class')
plt.ylabel('Number of Passengers')
plt.title('Number of Passengers by Class')
plt.show()
```

```python
# Adjust x-ticks to show labels
class_counts = titanic['survived'].value_counts()
print (class_counts)
class_counts.plot(kind='bar', color=["r","green"], edgecolor= "black" )
plt.xlabel('Alive/Dead')
plt.ylabel('Passenger Count')
plt.title('Survived VS Dead Passengers')
plt.xticks(ticks=[0, 1], labels=['Dead', 'Alive'])
lable=[plt.Line2D([0], [0], marker='s',c="w",label='dead',markerfacecolor='red',markersize=10),
       plt.Line2D([0], [0], marker='s',c="w",label='Alive',markerfacecolor='green',markersize=10)]
plt.legend(handles=lable)
plt.show()
```

# Pi Chart

```python
gender = titanic['sex'].value_counts()
gender.plot(kind='pie',autopct='%1.1f%%', colors=["purple","blue"])
lable1= [plt.Line2D([0],[0],marker="s",c='w',label='Male',markerfacecolor='purple',markeredgecolor='black',markersize=10),
         plt.Line2D([0],[0],marker="s",c="w",label='Female',markerfacecolor='blue',markeredgecolor='black',markersize=10)]
plt.legend(handles=lable1)
plt.show()
```
