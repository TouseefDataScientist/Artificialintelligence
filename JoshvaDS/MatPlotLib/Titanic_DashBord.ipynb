```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

titanic = sns.load_dataset('titanic')
plt.figure(figsize=(16,6))
plt.subplots_adjust(wspace=0.3, hspace=0.5) 

# Scatter Plot
plt.subplot( 2,2, 1)
x = titanic['age']
y = titanic['fare']
category = [plt.Line2D([0], [0], marker='s',
                       c="w",label='200 to Max',
                       markerfacecolor='purple',
                       markersize=10,
                       markeredgecolor='black'),
           plt.Line2D([0], [0], marker='s',
                       c="w",label='70 to 200',
                       markerfacecolor='blue',
                       markersize=10,
                       markeredgecolor='black'),
           plt.Line2D([0], [0], marker='s',
                      c="w",label='0 to 90',
                      markerfacecolor='royalblue',
                      markersize=10,
                      markeredgecolor='black')]

plt.legend(handles=category, title='Fare Categories')

plt.scatter(x, y,c=["purple" if x>200 else "blue" if x<70 else "royalblue" for x in y], edgecolor= "black" )
plt.title('Scatter Plot of Age vs Fare')


# Violin Plot
plt.subplot( 2,2, 2)
sns.violinplot(x='class', y='age', data=titanic, color='green',alpha=0.8)
plt.title('Age-wise Class Volume')

# Bar Plot
plt.subplot( 2,2, 3)
class_counts = titanic['survived'].value_counts().plot(kind='bar', color=["r","green"],alpha=0.8, edgecolor= "black" )
plt.title('Survived VS Dead Passengers')
plt.xticks(ticks=[0, 1], labels=['Dead', 'Alive'])
lable_bar=[plt.Line2D([0], [0], marker='s',c="w",label='dead',markerfacecolor='red',markersize=10),
       plt.Line2D([0], [0], marker='s',c="w",label='Alive',markerfacecolor='green',markersize=10)]
plt.legend(handles=lable_bar)
plt.title('Class distribution')

# Pie Chart
plt.subplot( 2,2, 4)
plt.pie(titanic['sex'].value_counts(),autopct='%1.1f%%', colors=["royalblue","pink"])
lable_pi=[plt.Line2D([0], [0], marker='s',c="w",label='Males',markerfacecolor='royalblue',markersize=10),
       plt.Line2D([0], [0], marker='s',c="w",label='Fe-males',markerfacecolor='pink',markersize=10)]
plt.legend(handles=lable_pi, loc='center right',bbox_to_anchor=(1, 0, .7, 1.5))
plt.title('Gender Persentage')

plt.suptitle('Titanic Data Visualization', fontsize=20)
plt.show()
```
