* A. A Vandermonde matrix is a matrix generated from a vector in which each column of the matrix is an integer power starting from zero. So, if I have a column vector [𝑥1,𝑥2,…,𝑥𝑁]𝑇 and then the associated (square) Vandermonde matrix would be: V= ......
Use what you learned about broadcasting in the previous problem to write a function that will produce a Vandermonde matrix for a vector [1,2,…,𝑁]𝑇[1,2,…,N] T for any 𝑁. Do it without using a loop. Here’s a stub to get you started:
def vandermonde(N):
    vec = np.arange(N) + 1
    Finish me.
Use your function for N=12, store it in a variable named vander, and print the result.

```python
def vandermonde(N):
    vec = np.arange(N) + 1 
    vander = vec[:, np.newaxis] ** np.arange(N)  
    return vander
# Use the function for N=12
vander = vandermonde(12)
print("Vandermonde Matrix for N=12:")
print(vander)
```

* B. Now, let’s make a pretend linear system problem with this matrix. Create a vector of all ones, of length 12, and call it x. Perform a matrix-vector multiplication of vander with the vector you just created and store that in a new vector and call it b. Print the vector 𝑏.

```python
# Create a vector of all ones of length 12
N = 12
x = np.ones(N)
# Perform matrix-vector multiplication
b = vander @ x 
#Printing Results:
print("Vector b:\n",b)
```

* C.  First, solve the linear system the naïve way, pretending like you don’t know x. Import numpy.linalg, invert V and multiply it by b. Print out your result. What should you get for your answer? If the answer is different than what you expected, write a sentence about that difference.

```python
#Inverting the Vandermonde matrix and multiply by b.
x_naive = np.linalg.inv(vander) @ b
#Printing the result.
print("Naive solution for x:\n",x_naive)
#Checking the difference from the original x.
difference = np.abs(x_naive - x)
print("Difference from expected x:",difference)
#"The naive method of inverting the Vandermonde matrix did not return the expected vector of ones, indicating that the inversion may have introduced numerical instability or inaccuracies due to the properties of the matrix."
```

* D. Now, solve the same linear system using solve. Print out the result. Does it seem more or less in line with what you’d expect?

```python
x_solve = np.linalg.solve(vander, b)
#Printing the result.
print("Solution using solve:", x_solve)
#Check if it matches the expected vector of ones
if np.allclose(x_solve, x):
    print("The solution is in line with what I expected: it matches the vector of ones.")  
else:
    print("The solution is not in line with what I expected.") 
```
