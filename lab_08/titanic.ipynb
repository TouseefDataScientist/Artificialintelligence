```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('datasets/train.csv')

df.head()
```

```python
x = df['Age']
```

```python
df.Age
```

```python
y = df['Fare']
```

```python
# plt.plot(x, y, marker='*')
plt.scatter(x, y, marker='*')

plt.xlabel('Age')
plt.ylabel('Fare')
plt.title('Age vs Fare')

plt.show()
```

```python
plt.scatter(x, y, color='red', marker='*')
plt.scatter(x, y, color='blue', marker='*', alpha=0.5)

plt.xlabel('Age')
plt.ylabel('Fare')
plt.title('Age vs Fare')

plt.show()
```

```python
plt.scatter(x, y, color='red', marker='+')
plt.scatter(x, y, color='blue', marker='*', alpha=0.5)

plt.xlabel('Age')
plt.ylabel('Fare')
plt.title('Age vs Fare')
plt.legend(['Age (red)', 'Fare (blue)'])

plt.show()
```
