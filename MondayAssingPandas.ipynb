#### Problem 01: “Advanced Pandas Data Manipulation with Kaggle Housing Dataset” 
A. Loading the CSV File

```python
housing_data = pd.read_csv("housing_price_dataset.csv")
housing_data
```

B. Filter houses priced above the median and those with more than 2 bedrooms and 2000 sqft living 
area. 

```python
median_price = housing_data['Price'].median()
filtered_houses = housing_data[(housing_data['Price'] > median_price) & 
                               (housing_data['Bedrooms'] > 2) & 
                               (housing_data['SquareFeet'] > 2000)]
filtered_houses
```

C. Group by bedrooms and calculate the average price and median living area. 

```python

grouped_by_bedrooms = filtered_houses.groupby('Bedrooms').agg(
    avg_price=('Price', 'mean'),
    median_living_area=('SquareFeet', 'median')
)

grouped_by_bedrooms
```

D. Sort by price and rank the top 5% most expensive houses. 

```python
housing_data = housing_data.sort_values(by='Price', ascending=False)

top_5_percent_count = int(len(housing_data) * 0.05)

top_5_percent = housing_data.head(top_5_percent_count)

print(top_5_percent) 
```

E. Calculate correlations between price, living area, and bedrooms. 

```python
subset = housing_data[['Price', 'SquareFeet', 'Bedrooms']]
correlation = subset.corr()
print(correlation)
```

F. Generate descriptive statistics for price, bedrooms, living area, and Year built. 

```python
descriptive_stats = housing_data[['Price', 'Bedrooms', 'SquareFeet', 'YearBuilt']].describe()

print(descriptive_stats)
```

G. Create a price_per_sqft column and classify houses as "Low," "Medium," or "High." 

```python
Prices_per_area=housing_data["Price"]/housing_data["SquareFeet"]
def classify_price_per_area(row):
    if row["price_per_area"]<100:
        return"Low"
    elif row["price_per_area"]<200:
        return"Medium"
    else:
        return"High"
Prices_per_area
```
