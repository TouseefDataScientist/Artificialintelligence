### ***Joins*** In Pandas

```python
left = pd.DataFrame({"key": ["foo", "foo"], "lval": [1, 2]})

right = pd.DataFrame({"key": ["foo", "foo"], "rval": [4, 5]})
# This code will merge two dataframes on the basis of a common key value
pd.merge(left,right,on="key")
```

```python
# Output will be different if we change the dataframe
left = pd.DataFrame({"key": ["one", "two"], "lval": [1, 2]})

right = pd.DataFrame({"key": ["one", "five"], "rval": [4, 5]})
pd.merge(left,right,on="key")
# Because the value of key is similar i.e one in only one column that why only one row is selected
```

```python
# Grouping of the dataframe
df_grouping = pd.DataFrame(
    {
        "A": ["foo", "bar", "foo", "bar", "foo", "bar", "foo", "foo"],
        "B": ["one", "one", "two", "three", "two", "two", "one", "three"],
        "C": np.random.randn(8),
        "D": np.random.randn(8),
    }
)
df_grouping
```

```python
# Grouping by a column label, selecting column labels, and then applying the DataFrameGroupBy.sum() function to the resulting groups:

df_grouping.groupby("A")[["C", "D"]].sum()
```

```python
# Grouping by multiple columns label forms MultiIndex.
df_grouping.groupby(["A", "B"]).sum()
```

#### Reshaping the data frame

```python
var=pd.DataFrame({"days":[1,2,3,4,5,6],"english":[10,12,14,12,18,20],"math":[17,18,19,13,14,16]})
var
```

```python
## How to make data look like Vertically
pd.melt(var)
```

```python
# Make a specific column as an ID
pd.melt(var,id_vars=["days"])
```

```python
# change the name of the header
pd.melt(var,id_vars=["days"],var_name="Subjects",value_name="Marks")
```

## Pivot Functoin

```python
varp=pd.DataFrame({"days":[1,2,3,4,5,6],"Name":["A","B","C","D","E","F"],"english":[10,12,14,12,18,20],"math":[17,18,19,13,14,16]})
varp
```

```python
varp.pivot(index="days",columns="Name",values="english")
```

```python
varp.pivot_table(index="Name",columns="days",aggfunc="sum",margins=True)
```
