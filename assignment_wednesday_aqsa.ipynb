#### Problem 01: “Advanced Plot Customization and Axes Control”

```python
import matplotlib.pyplot as plt
import numpy as np

# Create some data
x = np.linspace(1, 10, 100)
y1 = np.sin(x)
y2 = np.exp(x)

# Create the figure and the first y-axis
fig, ax1 = plt.subplots()

# Plot on the first y-axis with linear scale
ax1.plot(x, y1, 'g-')
ax1.set_xlabel('X-axis')
ax1.set_ylabel('Linear Scale', color='g')
ax1.tick_params(axis='y', labelcolor='g')

# Create the second y-axis with logarithmic scale
ax2 = ax1.twinx()
ax2.plot(x, y2, 'b-')
ax2.set_ylabel('Logarithmic Scale', color='b')
ax2.set_yscale('log')
ax2.tick_params(axis='y', labelcolor='b')

# Show gridlines for each y-axis
ax1.grid(True)
ax2.grid(True, which='both')  # for logarithmic scale

plt.title('Dual-Axis Plot with Linear and Logarithmic Scales')
plt.show()
```

#### Problem 02: “Interactive Plotting with Widgets”

```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.widgets import Slider

# Initial parameters
amp_init = 1
freq_init = 1
phase_init = 0

# Create the plot
x = np.linspace(0, 10, 1000)
y = amp_init * np.sin(freq_init * x + phase_init)

fig, ax = plt.subplots()
line, = ax.plot(x, y, lw=2)
ax.set_xlabel('x')
ax.set_ylabel('Amplitude')

# Adjust the subplots region to fit the sliders
plt.subplots_adjust(left=0.1, bottom=0.25)

# Define sliders
ax_amp = plt.axes([0.1, 0.1, 0.65, 0.03])
ax_freq = plt.axes([0.1, 0.15, 0.65, 0.03])
ax_phase = plt.axes([0.1, 0.2, 0.65, 0.03])

slider_amp = Slider(ax_amp, 'Amplitude', 0.1, 2.0, valinit=amp_init)
slider_freq = Slider(ax_freq, 'Frequency', 0.1, 2.0, valinit=freq_init)
slider_phase = Slider(ax_phase, 'Phase', 0, 2 * np.pi, valinit=phase_init)

# Update function
def update(val):
    amp = slider_amp.val
    freq = slider_freq.val
    phase = slider_phase.val
    line.set_ydata(amp * np.sin(freq * x + phase))
    fig.canvas.draw_idle()

# Attach update function to sliders
slider_amp.on_changed(update)
slider_freq.on_changed(update)
slider_phase.on_changed(update)

plt.show()
```

#### Problem 03: “3D Visualization with Custom Projections”

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np

# Generate data
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)
z = np.exp(-(x**2 + y**2))

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Customize 3D plot with a cylindrical projection
ax.plot_surface(x, y, z, cmap='viridis')
ax.set_box_aspect([1, 1, 0.5])  # for cylindrical appearance

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
plt.title('3D Gaussian Surface with Cylindrical Projection')
plt.show()
```


#### Problem 04: “Plot Efficiency and Performance Optimization”

```python
import matplotlib.pyplot as plt
import numpy as np
import time

# Generate large dataset
n = 10**6
x = np.random.randn(n)
y = np.random.randn(n)

# Downsampling
sample_rate = 0.1
x_sample = x[::int(1/sample_rate)]
y_sample = y[::int(1/sample_rate)]

# Start timer
start_time = time.time()

# Plot downsampled data
fig, ax = plt.subplots()
ax.scatter(x_sample, y_sample, alpha=0.1)
ax.set_title("Optimized Scatter Plot with Downsampling")

# End timer and print time taken
plt.show()
print(f"Time taken: {time.time() - start_time:.2f} seconds")
```

#### Problem 05: “Custom Colormaps and Data Normalization” 

```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.colors as mcolors

# Generate random data
data = np.random.rand(10, 10) * 100

# Normalize the data to emphasize specific ranges
norm = mcolors.TwoSlopeNorm(vmin=0, vcenter=50, vmax=100)

# Create a custom colormap emphasizing low (blue), mid (white), and high (red) values
cmap = mcolors.LinearSegmentedColormap.from_list('custom_cmap', ['blue', 'white', 'red'])

# Plot the heatmap
plt.imshow(data, cmap=cmap, norm=norm)
plt.colorbar()
plt.title('Heatmap with Custom Colormap and Normalization')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```

#### Problem 06: “Subplots with Shared Axes and Complex Layouts

```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.gridspec import GridSpec

# Generate sample data
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.tan(x)
y4 = np.exp(-x)

# Set up a figure and GridSpec layout
fig = plt.figure(figsize=(10, 8))
gs = GridSpec(2, 2, figure=fig, height_ratios=[1, 2], hspace=0.3, wspace=0.3)

# Define the subplots with shared axes
ax1 = fig.add_subplot(gs[0, 0])
ax2 = fig.add_subplot(gs[0, 1], sharey=ax1)
ax3 = fig.add_subplot(gs[1, :])  # This spans across both columns

# Plot data on subplots
c1 = ax1.scatter(x, y1, c=y1, cmap='viridis')
ax1.set_title('Sine Wave')
ax1.set_ylabel('Amplitude')

c2 = ax2.scatter(x, y2, c=y2, cmap='viridis')
ax2.set_title('Cosine Wave')

c3 = ax3.scatter(x, y3, c=y3, cmap='viridis')
ax3.set_title('Tangent Wave')
ax3.set_xlabel('X-axis')
ax3.set_ylabel('Amplitude')

# Add a shared color bar for the entire figure
cbar = fig.colorbar(c1, ax=[ax1, ax2, ax3], orientation='vertical', fraction=0.025, pad=0.04)
cbar.set_label('Color Intensity')

plt.show()

```
