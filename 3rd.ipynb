### *Problem 1: Dual-Axis Plot (Linear and Logarithmic)*



```python

import matplotlib.pyplot as plt
import numpy as np
```

# Generate sample data


```python
x = np.linspace(1, 10, 100)
y1 = np.exp(x)  
y2 = x ** 2  

```

# Create figure and first axis


```python
fig, ax1 = plt.subplots()
```

# Plot the first dataset with linear scale


```python
ax1.plot(x, y2, 'b-', label='Linear')
ax1.set_xlabel('X-axis')
ax1.set_ylabel('Linear Scale', color='b')
ax1.tick_params(axis='y', labelcolor='b')
ax1.grid(True)
```

# Create a second y-axis sharing the same x-axis


```python
ax2 = ax1.twinx()
ax2.plot(x, y1, 'r-', label='Logarithmic')
ax2.set_ylabel('Logarithmic Scale', color='r')
ax2.set_yscale('log')
ax2.tick_params(axis='y', labelcolor='r')

plt.title('Dual Axis Plot (Linear and Logarithmic)')
plt.show()
```

### *Problem 2: Interactive Plot with Widgets*


```python
import matplotlib.pyplot as plt
from matplotlib.widgets import Slider
import numpy as np
```

# Create figure and sine wave function


```python
fig, ax = plt.subplots()
plt.subplots_adjust(left=0.25, bottom=0.25)
t = np.linspace(0, 10, 1000)
initial_amp = 1
initial_freq = 1
initial_phase = 0
y = initial_amp * np.sin(initial_freq * t + initial_phase)
line, = plt.plot(t, y)
```


# Add sliders for amplitude, frequency, and phase


```python
axamp = plt.axes([0.25, 0.1, 0.65, 0.03])
axfreq = plt.axes([0.25, 0.15, 0.65, 0.03])
axphase = plt.axes([0.25, 0.2, 0.65, 0.03])

amp_slider = Slider(axamp, 'Amplitude', 0.1, 10.0, valinit=initial_amp)
freq_slider = Slider(axfreq, 'Frequency', 0.1, 10.0, valinit=initial_freq)
phase_slider = Slider(axphase, 'Phase', -np.pi, np.pi, valinit=initial_phase)
```

# Update function for sliders


```python
def update(val):
    amp = amp_slider.val
    freq = freq_slider.val
    phase = phase_slider.val
    line.set_ydata(amp * np.sin(freq * t + phase))
    fig.canvas.draw_idle()
```

# Register update function


```python
amp_slider.on_changed(update)
freq_slider.on_changed(update)
phase_slider.on_changed(update)

plt.show()
```

### *Problem 3: 3D Visualization with Custom Projections*


```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

```

# Define the function for 3D plot


```python
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)
z = np.exp(-0.1 * (x**2 + y**2))
```

# Create figure


```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Plot the surface
ax.plot_surface(x, y, z, cmap='viridis')

```

# Customize the projection


```python
ax.view_init(elev=30, azim=45)
plt.title('3D Gaussian Function with Custom Projection')
plt.show()
```

### *Problem 5: Custom Colormaps and Data Normalization*

### *Problem 6: Subplots with Shared Axes*


```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.gridspec import GridSpec

```

# Create sample data


```python
 x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.tan(x)
```

# Create a 2x2 grid of subplots using GridSpec


```python
fig = plt.figure(constrained_layout=True)
gs = GridSpec(3, 2, figure=fig)

ax1 = fig.add_subplot(gs[0, 0])
ax2 = fig.add_subplot(gs[0, 1])
ax3 = fig.add_subplot(gs[1, :])
ax4 = fig.add_subplot(gs[2, :])

```

# Plot data in subplots


```python
ax1.plot(x, y1)
ax2.plot(x, y2)
ax3.plot(x, y3)
```

# Shared axes


```python
ax4.plot(x, y1 + y2)
ax1.set_title('Sine Wave')
ax2.set_title('Cosine Wave')
ax3.set_title('Tangent Wave')
plt.show()
```

### *Problem 8: Advanced Visualization with Subplots and Annotations*


```python
import matplotlib.pyplot as plt
import numpy as np

```

# Generate time series data


```python
time = np.linspace(0, 365, 365)
temperature = 20 + 10 * np.sin(2 * np.pi * time / 365)
humidity = 50 + 20 * np.sin(2 * np.pi * time / 365 + np.pi / 4)

```

# Create subplots


```python
fig, ax1 = plt.subplots()
ax2 = ax1.twinx()
ax1.plot(time, temperature, 'g-', label='Temperature')
ax2.plot(time, humidity, 'b-', label='Humidity')

ax1.set_xlabel('Day of Year')
ax1.set_ylabel('Temperature (C)', color='g')
ax2.set_ylabel('Humidity (%)', color='b')

```

# Annotate significant events


```python
ax1.annotate('Summer Peak', xy=(182, 30), xytext=(200, 40), arrowprops=dict(facecolor='black', arrowstyle="->"))
ax1.annotate('Winter Dip', xy=(0, 10), xytext=(50, 15), arrowprops=dict(facecolor='black', arrowstyle="->"))
plt.title('Temperature and Humidity Over a Year')
plt.show()

```
