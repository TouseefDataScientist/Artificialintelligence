>Q.1: Sales Data Analysis: You have a NumPy array representing the daily sales of a product over a month. Each element in the array represents the sales for a particular day. Compute the total sales, average sales, and the day with the highest sales.

>Q.3: Financial Analysis: You have a NumPy array representing the daily closing prices of a stock for a month. Calculate the monthly return, which is the percentage change from the first to the last day of the month.

```python
closing_prices = np.array([150.0, 152.3, 153.4, 155.1, 156.7, 154.0, 157.2, 160.3, 161.0, 
                           162.4, 164.2, 162.9, 161.8, 163.5, 165.1, 166.0, 164.9, 163.8,
                           165.2, 167.3, 168.5, 169.2, 170.1, 172.0, 171.5, 173.2, 175.0, 
                           174.1, 176.8, 178.2])

first_day_price = closing_prices[0]
last_day_price = closing_prices[-1]

monthly_return = ((last_day_price - first_day_price) / first_day_price) * 100

print(f"Monthly return: {monthly_return:.2f}%")
```

>Q.6: Population Growth Analysis: You have a NumPy array representing the population of a city over 10 years. Calculate the annual growth rate and identify the year with the highest growth.


```python
population = np.array([10000, 10500, 11000, 11550, 12300, 13000, 13500, 14000, 15000, 16000])

annual_growth_rate = np.diff(population) / population[:-1] * 100

year_of_highest_growth = np.argmax(annual_growth_rate) + 1  

print("Annual Growth Rate (%):", annual_growth_rate)
print(f"The year with the highest growth is Year {year_of_highest_growth + 1} with a growth rate of {annual_growth_rate[year_of_highest_growth]:.2f}%")
```

>Q.7: Traffic Data Analysis: Given a 2D NumPy array where each row represents the number of vehicles passing through a checkpoint in different hours of the day over a week, calculate the total traffic for each day and the busiest hour of the week.


```python
traffic_data = np.array([
    [100, 120, 130, 140, 150, 160, 170, 180, 190, 200, 210, 220, 230, 240, 250, 260, 270, 280, 290, 300, 310, 320, 330, 340],
    [90, 100, 110, 120, 130, 140, 150, 160, 170, 180, 190, 200, 210, 220, 230, 240, 250, 260, 270, 280, 290, 300, 310, 320],
    [80, 90, 100, 110, 120, 130, 140, 150, 160, 170, 180, 190, 200, 210, 220, 230, 240, 250, 260, 270, 280, 290, 300, 310],
    [70, 80, 90, 100, 110, 120, 130, 140, 150, 160, 170, 180, 190, 200, 210, 220, 230, 240, 250, 260, 270, 280, 290, 300],
    [60, 70, 80, 90, 100, 110, 120, 130, 140, 150, 160, 170, 180, 190, 200, 210, 220, 230, 240, 250, 260, 270, 280, 290],
    [50, 60, 70, 80, 90, 100, 110, 120, 130, 140, 150, 160, 170, 180, 190, 200, 210, 220, 230, 240, 250, 260, 270, 280],
    [40, 50, 60, 70, 80, 90, 100, 110, 120, 130, 140, 150, 160, 170, 180, 190, 200, 210, 220, 230, 240, 250, 260, 270]
])

total_traffic_per_day = np.sum(traffic_data, axis=1)

busiest_hour_index = np.argmax(traffic_data)
busiest_hour_day = busiest_hour_index // traffic_data.shape[1]  # Row index (day)
busiest_hour_time = busiest_hour_index % traffic_data.shape[1]   # Column index (hour)

print("Total traffic for each day:", total_traffic_per_day)
print(f"Busiest hour of the week: Day {busiest_hour_day + 1}, Hour {busiest_hour_time + 1}, Vehicles: {traffic_data[busiest_hour_day, busiest_hour_time]}")
```

>Q.9: Game Score Analysis: You have a NumPy array representing the scores of different players in a game over multiple rounds. Identify the player with the highest average score and the round with the highest total score.

```python
scores = np.array([
    [10, 20, 30, 40],  # Player 1
    [15, 25, 35, 45],  # Player 2
    [10, 30, 50, 40],  # Player 3
    [20, 25, 35, 50]   # Player 4
])

average_score_per_player = np.mean(scores, axis=1)

total_score_per_round = np.sum(scores, axis=0)

player_with_highest_avg_score = np.argmax(average_score_per_player)

round_with_highest_total_score = np.argmax(total_score_per_round)

print("Average score per player:", average_score_per_player)
print(f"Player with the highest average score: Player {player_with_highest_avg_score + 1}")

print("Total score per round:", total_score_per_round)
print(f"Round with the highest total score: Round {round_with_highest_total_score + 1}")
```

>Q.10: Financial Portfolio Analysis: You have a NumPy array representing the daily returns of different assets in a financial portfolio over a month. Calculate the portfolio's overall daily return, and identify the asset with the highest volatility.

```python
returns = np.array([
    [0.01, 0.02, 0.015],  # Day 1
    [-0.005, 0.01, 0.02],  # Day 2
    [0.02, -0.01, 0.025],  # Day 3
    [0.015, 0.03, 0.02],   # Day 4
    [0.005, -0.015, 0.01], # Day 5
    [-0.01, 0.005, 0.03],  # Day 6
    [0.02, 0.01, -0.005],  # Day 7
    [0.01, -0.02, 0.015],  # Day 8
    [-0.01, 0.015, 0.01],  # Day 9
    [0.005, 0.025, 0.02],  # Day 10
    [0.03, -0.005, 0.01],  # Day 11
    [-0.005, 0.02, -0.01], # Day 12
    [0.025, 0.01, 0.015],  # Day 13
    [0.02, -0.015, 0.03],  # Day 14
    [0.005, 0.03, 0.025],  # Day 15
    [0.015, -0.005, 0.02], # Day 16
    [0.01, 0.015, 0.01],   # Day 17
    [-0.005, 0.02, -0.005],# Day 18
    [0.025, 0.01, 0.03],   # Day 19
    [-0.01, -0.02, 0.015], # Day 20
    [0.02, 0.03, 0.02],    # Day 21
    [-0.005, 0.02, 0.025], # Day 22
    [0.01, 0.01, -0.01],   # Day 23
    [0.015, 0.03, 0.03],   # Day 24
    [0.02, -0.01, 0.025],  # Day 25
    [0.005, 0.025, 0.01],  # Day 26
    [0.03, 0.01, 0.015],   # Day 27
    [-0.005, 0.02, -0.005],# Day 28
    [0.025, -0.01, 0.03],  # Day 29
    [0.02, 0.03, 0.015]    # Day 30
])

overall_daily_return = np.mean(returns, axis=1)

volatility_per_asset = np.std(returns, axis=0)

asset_with_highest_volatility = np.argmax(volatility_per_asset)

print("Overall daily return for the portfolio over 30 days:")
print(overall_daily_return)

print("\nVolatility (standard deviation) per asset:")
print(volatility_per_asset)

print(f"\nAsset with the highest volatility: Asset {asset_with_highest_volatility + 1}")
```
