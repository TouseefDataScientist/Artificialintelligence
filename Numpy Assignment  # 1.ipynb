### Q.1: Sales Data Analysis: 
### You have a NumPy array representing the daily sales of a product over a month. Each element in the array represents the sales for a particular day. Compute the total sales, average sales, and the day with the highest sales


```python
# Daily Sales dataframe for a month (30 days)

Monthly_Sales = np.random.randint(500,3000,30)

print("Monthly Sales:", Monthly_Sales)
```

```python
# Compute the total sales
total_sales = np.sum(Monthly_Sales)
print("Compute the total sales:", total_sales)

# Average sales 
average_sales = np.mean(Monthly_Sales)
print("Average Sales:", average_sales)

# day with the highest sales & Amount

day_with_highest_sales = np.argmax(Monthly_Sales) + 1
print("Day with Highest Sales:", day_with_highest_sales)

highest_sales_amount = np.max(Monthly_Sales)
print("Highest Sales Amount:",highest_sales_amount)
```

### Q.2: Weather Data Processing: 
### Given a 2D NumPy array where each row represents daily temperature readings for a week (7 days), find the average temperature for each day across all weeks

```python
# daily temperature readings 

daily_temp_reading = np.random.randint(22,45, size=(4,7))
print("Daily Temp Readings for Every Week:", daily_temp_reading)

# average temperature for each day across all weeks
average_temp  =  np.mean(daily_temp_reading, axis=0)
print("Average Temperature for 07 Day:", average_temp)

# Identify the highest temperature recorded and the corresponding day

highest_temperature = np.max(daily_temp_reading)
print("Highest Temperature:", highest_temperature) 

highest_temperature_day = np.unravel_index(np.argmax(daily_temp_reading), daily_temp_reading.shape)
print("Highest Temperature on Day:", highest_temperature_day[-1] + 1)
```

### Q.3: Financial Analysis: 
### You have a NumPy array representing the daily closing prices of a stock for a month. Calculate the monthly return, which is the percentage change from the first to the last day of the month

```python
# NumPy array representing the daily closing prices of a stock for a month.

stock_closing = np.random.randint(1000,8000,30)
print("Daily Closing Prices of a Stock For Month:", stock_closing)

# Monthly return: percentage change from the first to the last day

monthly_return = (stock_closing[-1] - stock_closing[0]) / stock_closing[0] * 100
print("Monthly Return (%):", monthly_return)

# daily percentage of changes throughout the month

daily_percentage_changes = np.diff(stock_closing) / stock_closing[:-1] * 100
print("Daily Percentage of Changes (%):", daily_percentage_changes)
```

### Q.4: Student Grades Analysis: 
### You have a 2D NumPy array where each row represents a student and each column represents their grades in different subjects. Calculate the average grade for each student and the overall class average for each subject

```python
# 2D array where each row represents a student and each column represents grades in different subjects.
student_grades = np.random.randint(40,100, size=(5,10))
print("Student Grades of Different Subjects:", student_grades)

# Average grade for each student.
average_grade = np.mean(student_grades, axis=1)
print("Average Grade for Each Student:", average_grade)

# Overall class average for each subject.
class_average_each_subject  = np.mean(student_grades, axis=0)
print("Class Average Each Subject:", class_average_each_subject)

# Identify the student with the highest average grade
highest_avg_student = np.argmax(average_grade) 
highest_avg_grade = np.max(average_grade) 
print("Student With Highest Average Grade:", highest_avg_student + 1, "With an Average Of:", highest_avg_grade)
```

### Q.5: Image Processing: 
### You have a grayscale image represented as a 2D NumPy array of pixel intensities. Normalize the pixel values to a range of 0 to 1.

```python
# Grayscale image (2D Numpy array) with pixel intensities ranging from 0 to 255.
grayscale_image = np.random.randint(0,255, size=(4,5))
print("Grayscale Image:", grayscale_image)

# Normalize the pixel values to a range of 0 to 1.
normalize_image = grayscale_image / 255.0
print("normalize_image:", normalize_image)

# Identify the pixel with the highest intensity and its location.
highest_intensity = np.max(grayscale_image)
print("Highest Intensity:", highest_intensity)

# Highest intensity Location. 
highest_intensity_location = np.unravel_index(np.argmax(grayscale_image), grayscale_image.shape)
print("Highest Intensity Location:", highest_intensity_location)
```

### Q.6: Population Growth Analysis
### You have a NumPy array representing the population of a city over 10 years.

```python
# Population of a city over 10 years.

city_population = np.array([15,10,12,18,20,27,32,38,40,43])
print("Population of a City Over 10 Years:", city_population)

# Annual growth rate for each year
annual_growth_rate = np.diff(city_population) / city_population[:-1] * 100
print("Annual Growth Rate for Each Year (%):", annual_growth_rate)

#Identify the year with the highest growth.
highest_growth_year_index = np.argmax(annual_growth_rate)
highest_growth_rate = annual_growth_rate[highest_growth_year_index]
print("Year with Highest Growth is:", highest_growth_year_index + 1, "with a growth rate of:", highest_growth_rate)

# Compute the average annual growth rate over the 10-year period.
average_growth_rate = np.mean(annual_growth_rate)
print("Average Annual Growth Rate Over 10-Years:", average_growth_rate)
```

### Q.7: Traffic Data Analysis\n",
### Given a 2D NumPy array where each row represents the number of vehicles passing through a checkpoint during different hours of the day over a week.Calculate the total traffic for each day and the busiest hour of the week.

```python
# 2D array representing traffic data for 1 Week 24*7 a Day.
traffic_data = np.random.randint(20,80, (7,24))
print("Daily vehicles passing through a checkpoint:", traffic_data)

# Total traffic for each day.
total_traffic_per_day = np.sum(traffic_data, axis=1)
print("Total traffic For Each Day:", total_traffic_per_day)

# Busiest hour of the week.
busiest_hour_index = np.argmax(np.sum(traffic_data, axis=0))
busiest_hour_traffic = np.sum(traffic_data, axis=0)[busiest_hour_index]
print("Busiest Hour of the Week:", busiest_hour_index, "with traffic of:", busiest_hour_traffic)
```
