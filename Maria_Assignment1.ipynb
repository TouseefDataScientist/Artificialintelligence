```python
import numpy as np


sale_data = np.array([100, 150, 200, 120, 180, 250, 170, 130, 190, 220])

t_sales = np.sum(sale_data)

average_sales = np.mean(sale_data)

highest_sales_day = np.argmax(sale_data) + 1 

print("Total Sales:", t_sales)
print("Average Daily Sales:", average_sales)
print("Day with Highest Sales:", highest_sales_day)
```
