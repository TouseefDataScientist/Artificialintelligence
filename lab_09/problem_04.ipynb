### **Purpose:** Problem 04: Plot Efficiency and Performance Optimization

```python
import numpy as np
import matplotlib.pyplot as plt
import time
```

```python
# Generate 1 million points
n = 1000000
x = np.random.rand(n)
y = np.random.rand(n)
```

```python
# Define a Gaussian function
def gaussian(x, y):
    return np.exp(-0.1*(x**2 + y**2))
```

```python
# Create grid data
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)
z = gaussian(x, y)
```

```python
# Measure the time taken for normal scatter plot
start_time = time.time()
plt.scatter(x, y, s=1, alpha=0.5)  # Reduce marker size and alpha for better performance
plt.title('Scatter Plot with 1 Million Points')
plt.show()
end_time = time.time()
print(f'Time taken: {end_time - start_time:.2f} seconds')
```

```python
# Experiment with downsampling (e.g., use only 10% of data)
downsampled_x = x[::10]
downsampled_y = y[::10]
```

```python
start_time = time.time()
plt.scatter(downsampled_x, downsampled_y, s=1, alpha=0.5)
plt.title('Downsampled Scatter Plot (10% of data)')
plt.show()
end_time = time.time()
print(f'Time taken with downsampling: {end_time - start_time:.2f} seconds')
```
