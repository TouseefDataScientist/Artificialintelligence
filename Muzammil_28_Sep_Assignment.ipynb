> > (A) You make a measurement of objects {1,2} against objects {3,4} and get a difference of 1 kg. Write this as a linear equation for the masses.

```python
import numpy as np

# Linear equation for Part A
m1, m2, m3, m4 = 0, 0, 0, 0  # Unknown masses initialized

# Equation based on the measurement
# Here, we can just express it as a simple function
def measurement_a(m1, m2, m3, m4):
    return m1 + m2 - m3 - m4 == 1

# Display the function call result
print("Measurement A result:", measurement_a(m1, m2, m3, m4))

```

> > (C) A very common use of broadcasting is to standardize data, i.e., to make it have zero mean and unit variance.
  First, create a fake “dataset” with 50 examples, each with five dimensions.
> >
> > import numpy.random as npr
> >
> > data = np.exp(npr . randn(50,5))
> >
> >You don’t worry too much about what this code is doing at this stage of the course, but for completeness: it
> >imports the NumPy random number generation library, then generates a 50×5 matrix of standard normal 
> >random variates and exponentiates them. The effect of this is to have a pretend data set of 50 independent
> >and identically-distributed vectors from a log-normal distribution.

```python
import numpy.random as npr
data = np.exp(npr.randn(50, 5))


mean = np.mean(data, axis=0)
std_dev = np.std(data, axis=0)

# Step 2: Standardize the dataset
standardized_data = (data - mean) / std_dev

print("\nStandardized dataset (zero mean, unit variance):\n", standardized_data)

```

> >(D) Now, compute the mean and standard deviation of each column. This should result in two vectors of length 5. You’ll need to think a little bit about how to use the axis argument to mean and std. Store these vectors in to variables and print both of them.

```python
column_means = np.mean(data, axis=0)
column_std_devs = np.std(data, axis=0)

# Print the results
print("Mean", column_means)
print()
print("Standard deviation", column_std_devs)
```

> > (E) Now standardize the data matrix by 
> > 1. subtracting the mean off of each column, and
> > 2. dividing each column by its standard deviation.
> > Do this via broadcasting, and store the result in a matrix called normalized.
> >
> > To verify that you successfully did it, compute the mean and standard deviation of the columns of normalized and print them out.

```python
#Standardizing
normalized = (data - column_means) / column_std_devs

#Verification
normalized_means = np.mean(normalized, axis=0)
normalized_std_devs = np.std(normalized, axis=0)

print("Mean of each column after normalization:\n", normalized_means)
print("\nStandard deviation of each column after normalization:\n", normalized_std_devs)
```

> > (A) Use what you learned about broadcasting in the previous problem to write a function that will produce a Vandermonde matrix for a vector[1,2,...,N] T for any N. Do it without using a loop. Here’s a stub to get you started:
 def vandermonde(N):
 vec = np.arange(N)+1
 Finish me.
 Use your function for N=12, store it in variable named vander, and print the result.


```python
import numpy as np

# Vandermonde matrix function
def vandermonde(N):
    vec = np.arange(N) + 1  # Create vector [1, 2, ..., N]
    vander = np.vander(vec, increasing=True)  # Generate Vandermonde matrix
    return vander

# Use N = 12
N = 12
vander = vandermonde(N)
print("Vandermonde matrix for N=12:")
print(vander)
```

 > > (B) Now, let’s make a pretend linear system problem with this matrix. Create a vector of all ones, of length 12, and call it x. Perform a matrix-vector multiplication of vander with the vector you just created and store that in a new vector and call it b. Print the vector b.

```python
# Create a vector of all ones, perform matrix-vector multiplication
x = np.ones(N)  # Vector of all ones
b = np.dot(vander, x)  # Matrix-vector multiplication
print("\nVector b (result of matrix-vector multiplication):")
print(b)
```

> > (C) First, solve the linear system the naïve way, pretending like you don’t know x. Import numpy.linalg, invert vander, and multiply it by b. Print out your result. What should you get for your answer? If the answer is different than what you expected, write a sentence about that difference.

```python
import numpy.linalg as la

# Part (C): Naive solution using matrix inversion
vander_inv = la.inv(vander)  # Invert the Vandermonde matrix
x_naive = np.dot(vander_inv, b)  # Multiply inverse by b
print("\nNaive solution using matrix inversion:")
print(x_naive)
```

> > (D) Now, solve the same linear system using the solve function from the numpy.linalg library (i.e., solve 𝐴𝑥=𝑏 Ax=b directly without inverting the matrix). Print out the result. Compare it to the result in Part (C) and write a sentence about any differences.

```python
# Part (D): Solving the system using numpy.linalg.solve
x_solve = la.solve(vander, b)  # Solve Ax = b without inverting
print("\nSolution using numpy.linalg.solve:")
print(x_solve)
```
