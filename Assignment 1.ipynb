```python
import numpy as np
foo=np.arange(0,30)
print(foo)
print("Shape of foo is :",foo.shape)
```

```python
bar=foo.reshape(5,6)
print(bar)
print("shape of bar is ",bar.shape)
```

```python
baz=foo.reshape(2,3,5)
print(baz)
print("Shape of baz is :",baz.shape)
```

```python
bar[1,0]= -1
print("bar after changing value:\n",bar)
print("foo after bar modification:\n",foo)
print("baz after foo modification:\n ",baz)
```

```python
sd2=np.sum(baz,axis=1)
print("Sum of baz over second dimension is :",sd2)
```

```python
sd3=np.sum(baz,axis=2)
print("Sum of baz over third dimension is :",sd3)
```

```python
sd13=np.sum(baz,axis=(0,2))
print("Sum of baz over first and third dimension is :",sd13)
```

```python
import numpy.random as npr
data=np.exp(npr.randn(50,5))
mean=np.mean(data,axis=0)
print("Mean :",mean)
SD=np.std(data,axis=0)
print("Standard Deviation:",SD)
```

```python
normalized=(data-mean)/SD
normalized_Mean=np.mean(normalized,axis=0)
normalized_SD=np.std(normalized,axis=0)
print("Mean of each coloumn in normalized data:",normalized_Mean)
print("Standard deviation of each coloumn in normalized data:",normalized_SD)
```
