(A) You make a measurement of objects {1, 2} against objects {3, 4} and get a difference of 1kg. Write this as a linear equation for the masses.

```python
weight = np.array([[2, 1, -7, -1, 10]])
fstr1 = np.array([1])
print("Coefficient matrix A:\n",weight)
print("\nRight-hand side vector b:\n",fstr1)
```

(B) What is the minimum number of such measurements necessary to recover all five masses?

```python
weight1 = np.array([[1, 1, -9, 10, 0],[1, -1, 1, 0, 9],[0, 1, 1, -2, -1],[1, 0, 1, 44, 1],[-1, 1, 10, 1, -1]])
all_r = np.array([1, 2, 3, -2, 0])
print("Coefficient matrix Waight experiments:\n",weight1)
print("\nRight-hand side vector b:\n",all_r)
print("\nMinimum number of measurements needed to recover all masses:", weight1.shape[0])
```

(C) If the answer to the above question is N, come up with N − 1 measurements that, when combined with the answer to (A) above would make it possible to recover the masses.


```python
print("\nFirst equation coefficients:", weight1)
print("First equation result:", fstr1)
```

(D) What would A and b be in the matrix form of the linear system corresponding to the N measurements?
That is, if you were to set this up as Am = b where m is the vector of unknown masses. Denote the result of the ith measurement as bi . (Typeset these using the bmatrix environment.)

```python
mass = np.array(['m1', 'm2', 'm3', 'm4', 'm5'])
print("\nFinal system of equations:")
for i in range (all_r.size):
    equation_parts = []
    for j in range(weight.size):
        if weight1[i][j] != 0:
            equation_parts.append(f"+({weight1[i][j]}*{mass[j]})")
    print(" ".join(equation_parts), "=", all_r[i])
```

import numpy.random as npr

The effect of this is to have a pretend data set of 50 independent and identically-distributed vectors from a log-normal distribution.

```python
d_m = data.mean(axis=0)
d_std = data.std(axis=0)
print("Mean of each column:\n",d_m)
print("Standard deviation of each column:\n",d_std)
```
