## ***Linear Algebra in NumPy***
NumPy provides a set of functions for linear algebra, which are part of numpy.linalg.

1. Matrix Multiplication (np.dot() or @)
Matrix multiplication between two arrays can be done using np.dot() or @.

```python
arr2t = arry2.T
multiplication = np.dot(arry1,arr2t)
multiplication
```

Determinant (np.linalg.det())
The determinant of a matrix.

```python

square_matrix = arry3[0, :3, :3]
det_A = np.linalg.det(square_matrix)
det_A
```

```python
inv_A = np.linalg.inv(square_matrix)
inv_A
```

```python
eigenvalues= np.linalg.eig(square_matrix)
eigenvectors = np.linalg.eig(square_matrix)
print(eigenvalues)
print(eigenvectors)
```
