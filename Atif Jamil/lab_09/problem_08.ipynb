### **Purpose:** Problem 08: Advanced Visualization Techniques with Subplots and Annotations

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Generate some time-series data (1 year of data)
days = np.linspace(0, 365, 365)
temperature = 20 + 10 * np.sin(2 * np.pi * days / 365)  # Simulating seasonal temperature variation
humidity = 60 + 20 * np.cos(2 * np.pi * days / 365)  # Simulating seasonal humidity variation
```

```python
# Create subplots with shared x-axis
fig, (ax1, ax2) = plt.subplots(2, 1, sharex=True, figsize=(10, 6))

# Plot temperature data on the first subplot
ax1.plot(days, temperature, label='Temperature', color='r')
ax1.set_ylabel('Temperature (°C)')
ax1.set_title('Temperature and Humidity Over a Year')

# Highlight significant changes in temperature (summer and winter)
ax1.annotate('Summer peak', xy=(180, 30), xytext=(200, 35),
             arrowprops=dict(facecolor='black', shrink=0.05))
ax1.annotate('Winter low', xy=(0, 10), xytext=(30, 15),
             arrowprops=dict(facecolor='black', shrink=0.05))

# Plot humidity data on the second subplot
ax2.plot(days, humidity, label='Humidity', color='b')
ax2.set_xlabel('Days')
ax2.set_ylabel('Humidity (%)')

# Highlight significant changes in humidity (dry and humid seasons)
ax2.annotate('Humid peak', xy=(90, 80), xytext=(120, 85),
             arrowprops=dict(facecolor='black', shrink=0.05))
ax2.annotate('Dry season', xy=(270, 40), xytext=(300, 45),
             arrowprops=dict(facecolor='black', shrink=0.05))

# Set common x-axis label
fig.text(0.5, 0.04, 'Day of the Year', ha='center')

# Add grid lines for better visualization
ax1.grid(True)
ax2.grid(True)

# Show the plot
plt.tight_layout()
plt.show()
```
