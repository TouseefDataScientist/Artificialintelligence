### **Purpose:** Problem 02: Interactive Plotting with Widgets

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.widgets import Slider
```

```python
# Initial parameters
amp_init = 1
freq_init = 1
phase_init = 0
```

```python
# Define the sine function
def update(val):
    amp = amp_slider.val
    freq = freq_slider.val
    phase = phase_slider.val
    y = amp * np.sin(2 * np.pi * freq * x + phase)
    line.set_ydata(y)
    fig.canvas.draw_idle()
```

```python
# Create x data
x = np.linspace(0, 2 * np.pi, 1000)
y = amp_init * np.sin(2 * np.pi * freq_init * x + phase_init)
```

```python
# Create the plot
fig, ax = plt.subplots()
plt.subplots_adjust(left=0.1, bottom=0.3)
line, = ax.plot(x, y, lw=2)
```

```python
# Add sliders for amplitude, frequency, and phase
ax_amp = plt.axes([0.1, 0.2, 0.65, 0.03])
ax_freq = plt.axes([0.1, 0.15, 0.65, 0.03])
ax_phase = plt.axes([0.1, 0.1, 0.65, 0.03])
```

```python
amp_slider = Slider(ax_amp, 'Amplitude', 0.1, 5.0, valinit=amp_init)
freq_slider = Slider(ax_freq, 'Frequency', 0.1, 5.0, valinit=freq_init)
phase_slider = Slider(ax_phase, 'Phase', 0, 2 * np.pi, valinit=phase_init)
```

```python
# Update the plot when sliders change
amp_slider.on_changed(update)
freq_slider.on_changed(update)
phase_slider.on_changed(update)

plt.show()
```
