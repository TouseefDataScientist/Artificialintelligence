### **Purpose:** Problem 01: Advanced Plot Customization and Axes Control

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Create some data
x = np.linspace(0.1, 10, 100)
y1 = np.sin(x)  # Linear scale data
y2 = np.log(x)  # Logarithmic scale data
```

```python
# Create figure and axis
fig, ax1 = plt.subplots()

# Plot data on the first axis
ax1.plot(x, y1, 'b-', label='Sine (linear scale)')
ax1.set_xlabel('X axis')
ax1.set_ylabel('Sine', color='b')
ax1.tick_params('y', colors='b')
ax1.grid(True)

# Create a second y-axis sharing the same x-axis
ax2 = ax1.twinx()
ax2.plot(x, y2, 'r--', label='Logarithm (log scale)')
ax2.set_ylabel('Logarithm', color='r')
ax2.tick_params('y', colors='r')
ax2.set_yscale('log')  # Set logarithmic scale

plt.title('Dual-Axis Plot: Sine and Logarithm')
plt.show()
```
