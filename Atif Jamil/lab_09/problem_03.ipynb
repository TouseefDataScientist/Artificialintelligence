### **Purpose:** Problem 03: 3D Visualization with Custom Projections

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Define a Gaussian function
def gaussian(x, y):
    return np.exp(-0.1*(x**2 + y**2))
```

```python
# Create grid data
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)
z = gaussian(x, y)
```

```python
# Create figure and 3D axis
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Plot surface with color map
surf = ax.plot_surface(x, y, z, cmap='viridis')

# Add color bar
fig.colorbar(surf)

# Customize projection (you can experiment with different projections)
ax.view_init(elev=30, azim=120)

plt.show()
```
