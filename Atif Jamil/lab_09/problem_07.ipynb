### **Purpose:** Problem 07: Interactive Data Visualization with Matplotlib and Widgets

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.widgets import Slider
```

```python
# Define the sine wave function for plotting
def plot_sine_wave(amplitude, frequency):
    # Create the x data (time or angle in radians)
    x = np.linspace(0, 2 * np.pi, 1000)
    # Calculate the y data based on the sine function
    y = amplitude * np.sin(frequency * x)
    # Plot the sine wave
    line.set_ydata(y)
    fig.canvas.draw_idle()
```

```python
# Set initial parameters for the sine wave
initial_amplitude = 1
initial_frequency = 1
```

```python
# Create the x data (time or angle in radians)
x = np.linspace(0, 2 * np.pi, 1000)
# Calculate the initial y data based on the sine function
y = initial_amplitude * np.sin(initial_frequency * x)
```

```python
# Create the figure and axis for plotting
fig, ax = plt.subplots()
plt.subplots_adjust(left=0.1, bottom=0.3)
line, = ax.plot(x, y, lw=2)
ax.set_xlabel('x (radians)')
ax.set_ylabel('Amplitude')

# Create the sliders for interactive control
# Slider for adjusting amplitude
ax_amp = plt.axes([0.1, 0.2, 0.65, 0.03])
amp_slider = Slider(ax_amp, 'Amplitude', 0.1, 5.0, valinit=initial_amplitude)

# Slider for adjusting frequency
ax_freq = plt.axes([0.1, 0.1, 0.65, 0.03])
freq_slider = Slider(ax_freq, 'Frequency', 0.1, 5.0, valinit=initial_frequency)

# Update the plot based on the slider values
def update(val):
    # Get the current values of the sliders
    amp = amp_slider.val
    freq = freq_slider.val
    # Update the plot with the new amplitude and frequency
    plot_sine_wave(amp, freq)

# Connect the update function to the slider changes
amp_slider.on_changed(update)
freq_slider.on_changed(update)

# Show the plot with interactive sliders
plt.show()
```
