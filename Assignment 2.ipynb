### Game score analysis

### Financial Portfolio AnaLysis

```python
import numpy as np


daily_returns = np.array([
    [0.01, 0.02, -0.01, 0.03, 0.02], 
    [0.03, 0.01, 0.02, -0.01, 0.04],  
    [-0.02, 0.04, 0.03, 0.01, -0.01], 
    [0.01, -0.02, 0.02, 0.02, 0.03]    
])


odr = np.sum(daily_returns, axis=0)
print("Overall daily returns of the portfolio:", odr)

volatility = np.std(daily_returns, axis=1)
hvi = np.argmax(volatility)
hv = volatility[hvi]
print("Asset with the highest volatility: ",hvi + 1 ,"with a volatility of",hv)


hrdi = np.argmax(odr)
hrv = odr[hrdi]
print(f"Day with the highest portfolio return: Day {hrdi + 1} with a return of {hrv:.4f}")
```

### Health Care Data Analysis

```python
from numpy import random
patients_treated=np.random.randint(0,100,size=365)
tp=np.sum(patients_treated)
print("Total patients treated in whole year:",tp)
avg_p=np.mean(patients_treated)
print("Average patients treated per day:",avg_p)
hpc=np.max(patients_treated)
hpi=np.argmax(patients_treated)
print("Highest number of patients treated:",hpc, "on Day",hpi+1)

dav= np.sum(patients_treated > avg_p)
print("Number of days where treated patients exceeded the average daily count:", dav)


days_fewer_than_50 = np.sum(patients_treated < 50)
percentage_fewer_than_50 = (days_fewer_than_50 / len(patients_treated)) * 100
print("Percentage of days where the hospital treated fewer than 50 patients:", percentage_fewer_than_50)
```
