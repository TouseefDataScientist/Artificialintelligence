<H3>Sales Data Analysis

```python
sale_arr = np.array([1200,1300,1400,1500,1600,1700,1500])
print("Toatal sales OF the Day:", sale_arr)
print("Total sales :", np.sum(sale_arr) )
print("Average sales :", np.sum(sale_arr) )
print("Highest sales :", np.max(sale_arr)  )
highest_index = sale_arr.argmax()
print("Day of Highest Sale:", highest_index + 1)  
```

<h3>Weather Data Processing

```python
temperature_data = np.array([
    [72, 75, 78, 76, 73, 74, 75],  
    [68, 70, 72, 71, 69, 70, 71],  
    [65, 67, 70, 68, 66, 67, 68],  
    [80, 82, 85, 83, 81, 82, 83], 
    [77, 79, 82, 80, 78, 79, 80],  
    [85, 87, 90, 88, 86, 87, 88],  
    [82, 84, 87, 85, 83, 84, 85]  
])

average_temp_per_day = np.mean(temperature_data, axis=1)
print("Average temperature of the Day :", average_temp_per_day  )
max_temp = np.max(temperature_data)
print("Maximum temperature of the Day :", max_temp )
day_of_max_temp = np.argmax(np.max(temperature_data, axis=1)) + 1  
print("Highest temperature of the Day :", day_of_max_temp  )
```

<h3>Financial Analysis

```python
prices = np.array([100, 102, 101, 104, 103, 105, 106, 104, 107, 110,
                           109, 111, 113, 115, 114, 116, 117, 118, 120, 121,
                           123, 122, 124, 125, 128, 127, 126, 130, 132, 135])

firstday = prices[0]
lastday = prices[-1]
monthly_return = (((lastday - firstday) / firstday )* 100)
print("Mounthly Return :", monthly_return)
dailyprec = np.diff(prices) / prices[:-1] * 100
print("daily  percentage:", dailyprec)

```

<h3>Student Grades Analysis

```python
grades = np.array([[85, 90, 78], 
                   [92, 88, 84], 
                   [75, 85, 80], 
                   [89, 91, 87]])

student_averages = np.mean(grades, axis=1)
print("Average grade for each student:", student_averages)
subject_averages = np.mean(grades, axis=0)
print("Class average for each subject:", subject_averages)
highest_avg_student_idx = np.argmax(student_averages)
print("Student with the highest average grade (index):", highest_avg_student_idx)
highest_avg_student = student_averages[highest_avg_student_idx]
print("Student with the highest average grade (index):", highest_avg_student_idx)

```

<h3> Population Growth Analysis

```python
population = np.array([50000, 52000, 54000, 58000, 61000, 63000, 67000, 69000, 71000, 75000])
growth_rate = (population[1:] - population[:-1]) / population[:-1] * 100
print("Annual growth rates:", growth_rate)
highest_growth_year_idx = np.argmax(growth_rate) + 1
print("Year with the highest growth:", highest_growth_year_idx)
highest_growth_rate = growth_rate[highest_growth_year_idx - 1]
print("Highest growth rate (%):", highest_growth_rate)
average_growth_rate = np.mean(growth_rate)
print("Average annual growth rate (%):", average_growth_rate)
```

<h3>Traffic Data Analysis

```python
traffic_data = np.array([
    [120, 135, 150, 170, 160, 180, 200, 210, 230 ], 
    [130, 140, 160, 150, 155, 165, 185, 215, 225 ], 
    [125, 155, 165, 180, 170, 175, 200, 220, 240 ],
    [140, 150, 170, 190, 180, 175, 205, 215, 230 ], 
    [135, 145, 155, 165, 175, 185, 195, 205, 215], 
    [150, 160, 170, 180, 190, 200, 210, 220, 230],  
    [155, 165, 175, 185, 195, 205, 215, 225, 235]
])

total_traffic_per_day = np.sum(traffic_data, axis=1)
print("Total traffic for each day:", total_traffic_per_day)
busiest_hour = np.argmax(np.sum(traffic_data, axis=0))
print("Busiest hour of the week (0-23):", busiest_hour)
busiest_hour_vehicles = np.max(np.sum(traffic_data, axis=0))
print("Number of vehicles passing through at the busiest hour:", busiest_hour_vehicles)
```
