# <span style="color:violet"> <u>**User defined functions**</u> </span>
***DataFrame.agg()*** and ***DataFrame.transform()*** applies a user defined function that reduces or broadcasts its result respectively.

```python
df1 = pd.DataFrame({
    'A': [1, 2, 3],
    'B': [4, 5, 6]
})
def multiply_mean(x):
    return np.mean(x) * 100
df1.agg(multiply_mean).head()
```

# <span style="color:orange"><u>***Grouping***</u></span>
***group by*** we are referring to a process involving one or more of the following steps

**Splitting** the data into groups based on some criteria

**Applying** a function to each group independently

**Combining** the results into a data structure

This module contains the Grouping by multiple columns label forms


```python
df.groupby("Gender")[["Salary","Age"]].sum()
```

```python
df.groupby(["Age","Salary"]).sum().iloc[:,[0,2,3,5]]
```
