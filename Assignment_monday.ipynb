a. Create a DataFrame with synthetic data for students' grades, attendance, and extracurricular

```python


np.random.seed(42)  
num_students = 100
grades = np.random.randint(50, 101, size=(num_students, 5))  
attendance = np.random.uniform(70, 100, size=num_students)  
extracurricular_activities = np.random.choice(['Sports', 'Music', 'Art', 'None'], size=num_students)


students_df = pd.DataFrame({
    'student_id': range(1, num_students + 1),
    'grades': list(grades),
    'attendance': attendance,
    'extracurricular_activity': extracurricular_activities
})


students_df['average_grade'] = students_df['grades'].apply(np.mean)


filtered_students = students_df[
    (students_df['average_grade'] > 85) & 
    (students_df['attendance'] >= 90) |
    (students_df['grades'].apply(lambda x: any(g < 60 for g in x)) & (students_df['attendance'] < 90))
]

print(filtered_students)
```


b. Group the DataFrame by extracurricular_activity to calculate average grades and total attendance for each activity, then identify which activity has the highest average grade.


```python

grouped_df = students_df.groupby('extracurricular_activity').agg(
    average_grade=('average_grade', 'mean'),
    total_attendance=('attendance', 'sum')
).reset_index()


highest_avg_grade_activity = grouped_df.loc[grouped_df['average_grade'].idxmax()]

print(grouped_df)
print("Activity with highest average grade:", highest_avg_grade_activity)
```


c. Sort the DataFrame by grades and create a pass_fail column using a lambda function to indicate
passing or failing status, then count the total number of passes and fails


```python


students_df['pass_fail'] = students_df['average_grade'].apply(lambda x: 'Pass' if x >= 60 else 'Fail')
pass_fail_counts = students_df['pass_fail'].value_counts()

print(students_df.sort_values(by='average_grade'))
print("Pass counts:", pass_fail_counts['Pass'], "Fail counts:", pass_fail_counts['Fail'])
```


d. Create a second DataFrame containing parent income levels and merge it with the first DataFrame; analyze how these income levels impact the average grades of students.


```python

income_levels = np.random.choice(['Low', 'Medium', 'High'], size=num_students)
income_df = pd.DataFrame({
    'student_id': range(1, num_students + 1),
    'parent_income_level': income_levels
})


merged_df = pd.merge(students_df, income_df, on='student_id')


income_analysis = merged_df.groupby('parent_income_level')['average_grade'].mean().reset_index()

print(income_analysis)
```
