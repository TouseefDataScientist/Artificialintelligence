### 6.	Population Growth Analysis

```python
population_of_city=np.array([10500,12000,12500,142000,52300,62010,45260,48920,86321,78943])
population_of_city
```

```python
Difference= np.diff(population_of_city)
```

```python
last_population= population_of_city[-1]
```

```python
annual_growth_rate = Difference / last_population * 100
print(f"Annual Growth Rate is \n: {annual_growth_rate}")
```

```python
maximum_growth_index = np.argmax(annual_growth_rate)
maximum_growth_index
```

```python
maximum_growth_rate = annual_growth_rate[maximum_growth_index]
maximum_growth_rate
```

```python
year_of_maximum_growth = maximum_growth_index + 1  
print(f"Annual Maximum Growth is \n: {year_of_maximum_growth}")
```

```python
higest_average=np.mean(annual_growth_rate)
higest_average
```
